```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load Data

```python
#loading log data of the gru
own_train = pd.read_csv('results/cnn_train_own_embedding_trainable_training_csv.log')
own_untrain = pd.read_csv('results/cnn_train_own_embedding_untrainable_training_csv.log')
glove_train = pd.read_csv('results/cnn_cnn_train_glove_embedding_trainable_training_csv.log')
glove_untrain = pd.read_csv('results/cnn_train_glove_embedding_untrainable_training_csv.log')
no_pre = pd.read_csv('results/cnn_no_textcleaning_training_csv.log')
own_train.head()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 12))

axes[0].plot(own_train['epoch'],own_train['val_acc'], 'b',label = 'trainable_val_acc')
axes[0].plot(own_train['epoch'],own_train['acc'], 'g',label = 'trainable_acc')

axes[0].plot(own_untrain['epoch'],own_untrain['val_acc'], 'y',label = 'untrainable_val_acc')
axes[0].plot(own_untrain['epoch'],own_untrain['acc'], 'r',label = 'untrainable_acc')

axes[0].set_title("Own Embedding")
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Accuracy')
axes[0].legend()

axes[1].plot(glove_train['epoch'],glove_train['val_acc'], 'b',label = 'trainable_val_acc')
axes[1].plot(glove_train['epoch'],glove_train['acc'], 'g',label = 'trainable_acc')

axes[1].plot(glove_untrain['epoch'],glove_untrain['val_acc'], 'y',label = 'untrainable_val_acc')
axes[1].plot(glove_untrain['epoch'],glove_untrain['acc'], 'r',label = 'untrainable_acc')

axes[1].set_title("Glove Embedding")
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Accuracy')
axes[1].legend()

axes[2].plot(no_pre['epoch'],no_pre['val_acc'], 'b',label = 'val_acc')
axes[2].plot(no_pre['epoch'],no_pre['acc'], 'g',label = 'acc')

axes[2].set_title("No Pre-Trained Embedding / Text Cleaning")
axes[2].set_xlabel('Epochs')
axes[2].set_ylabel('Accuracy')
axes[2].legend()
plt.tight_layout()
```

```python
pre_train = pd.read_csv('results/vgg16_pretrained_trainable_training_csv.log')
un_train = pd.read_csv('results/vgg16_untrained_trainable_training_csv.log')
pre_untrain = pd.read_csv('results/vgg16_pretrained_untrainable_training_csv.log')
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)


axes.plot(pre_train['epoch'],pre_train['val_acc'], 'b',label = 'pretrained_trainable_val_acc')
axes.plot(pre_train['epoch'],pre_train['acc'], 'g',label = 'pretrained_trainable_acc')

axes.plot(un_train['epoch'],un_train['val_acc'], 'y',label = 'untrained_trainable_val_acc')
axes.plot(un_train['epoch'],un_train['acc'], 'r',label = 'untrained_trainable_acc')

axes.plot(pre_untrain['epoch'],pre_untrain['val_acc'], 'y',label = 'pretrained_untrainable_val_acc')
axes.plot(pre_untrain['epoch'],pre_untrain['acc'], 'r',label = 'pretrained_untrainable_acc')

axes.set_title("Transfer learning with VGG16 model")
axes.set_xlabel('Epochs')
axes.set_ylabel('Accuracy')
axes.legend()
plt.tight_layout()
```
